```python
import matplotlib.pyplot as plt
import numpy as np

# Nmero de sprints
sprints = np.arange(1, 7)

# Datos simulados para cada mtrica
tasa_exito_envio = [60, 70, 75, 85, 90, 95]  # En porcentaje
errores_validacion = [15, 12, 10, 8, 5, 2]  # Cantidad de errores
verificacion_layout = [6, 7, 7.5, 8, 8.5, 9]  # Escala de 1 a 10
verificacion_funcionalidad = [5, 6, 7, 8, 9, 9.5]  # Escala de 1 a 10
verificacion_compatibilidad = [4, 5, 6.5, 7, 8, 9]  # Escala de 1 a 10
documentacion_resultados = [50, 60, 65, 75, 85, 90]  # En porcentaje
satisfaccion_usuario = [5, 6, 7, 8, 9, 10]  # Escala de 1 a 10

# Configurar la figura para las grficas
plt.figure(figsize=(14, 10))

# Tasa de xito de Envo
plt.subplot(3, 2, 1)
plt.plot(sprints, tasa_exito_envio, marker='o', color='b')
plt.title("Tasa de xito de Envo (%)")
plt.xlabel("Sprint")
plt.ylabel("Porcentaje")
plt.ylim(0, 100)

# Nmero de Errores de Validacin
plt.subplot(3, 2, 2)
plt.plot(sprints, errores_validacion, marker='o', color='r')
plt.title("Nmero de Errores de Validacin")
plt.xlabel("Sprint")
plt.ylabel("Errores")
plt.ylim(0, 20)

# Verificacin de Layout y Diseo
plt.subplot(3, 2, 3)
plt.plot(sprints, verificacion_layout, marker='o', color='g')
plt.title("Verificacin de Layout y Diseo")
plt.xlabel("Sprint")
plt.ylabel("Puntuacin")
plt.ylim(0, 10)

# Verificacin de Funcionalidad
plt.subplot(3, 2, 4)
plt.plot(sprints, verificacion_funcionalidad, marker='o', color='m')
plt.title("Verificacin de Funcionalidad")
plt.xlabel("Sprint")
plt.ylabel("Puntuacin")
plt.ylim(0, 10)

# Verificacin de Compatibilidad
plt.subplot(3, 2, 5)
plt.plot(sprints, verificacion_compatibilidad, marker='o', color='c')
plt.title("Verificacin de Compatibilidad")
plt.xlabel("Sprint")
plt.ylabel("Puntuacin")
plt.ylim(0, 10)

# Documentacin de Resultados
plt.subplot(3, 2, 6)
plt.plot(sprints, documentacion_resultados, marker='o', color='y')
plt.title("Documentacin de Resultados (%)")
plt.xlabel("Sprint")
plt.ylabel("Porcentaje")
plt.ylim(0, 100)

plt.tight_layout()
plt.show()
```
